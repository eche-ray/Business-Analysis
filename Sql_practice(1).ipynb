```python
import pandas as pd
import sqlite3
```

```python
conn=sqlite3.connect('chinook.db')
cursor=conn.cursor()
```

```python
cursor.execute("""SELECT name FROM sqlite_master WHERE type='table';""")
tables=cursor.fetchall()
print(tables)
```

SELECT + WHERE + ORDER BY

```python
#Retrieve all customers from USA
df=pd.read_sql_query("""SELECT * FROM customers WHERE Country= 'USA';""",conn)
df.head()
```

```python
#Show employees ordered by hire date (newest first).
df=pd.read_sql_query("""SELECT EmployeeId, LastName, FirstName, HireDate FROM Employees ORDER BY HireDate DESC;""",conn)
df.head()
```

```python
#Find invoices with totals greater than 15.
df=pd.read_sql_query("""SELECT * FROM Invoices WHERE Total>15;""",conn)
df.head()

```

```python
#Display tracks longer than 5 minutes.
df=pd.read_sql_query("""SELECT * FROM invoice_items;""",conn)
df.head()

```

```python
#Retrieve customers whose first name starts with “A”.
df=pd.read_sql_query("""SELECT FirstName FROM Customers WHERE FirstName LIKE 'A%';""",conn)
df.head()

```

AND OR IN & LIKE

```python
#Find customers from USA or Canada.
df=pd.read_sql_query("""SELECT * FROM Customers WHERE Country= 'Canada' OR 'USA';""",conn)
df.head()

```

```python
#Find invoices between $5 and $15.

df=pd.read_sql_query("""SELECT * FROM invoices WHERE Total BETWEEN 5 AND 15;""",conn)
df.head()
```

```python
#Retrieve employees with titles containing “Manager
df=pd.read_sql_query("""SELECT Title FROM Employees WHERE Title LIKE '%Manager';""",conn)
df.head()
```

```python
#Find customers whose country is IN ('Brazil', 'Germany', 'France').
df=pd.read_sql_query("""SELECT * FROM Customers WHERE Country IN ('Brazil','Germany','France');""",conn)
df.head()
```
